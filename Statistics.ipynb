Play around with statistic, data visualization using dataset 'Wine quality' from UCI ML Repository.

Wine Quality data set is taken from https://archive.ics.uci.edu/ml/datasets/Wine+Quality

**Abstract:**
Two datasets are included, related to red and white vinho verde wine samples, from the north of Portugal. The goal is to model wine quality based on physicochemical tests 

**Data Set Information:**

The two datasets are related to red and white variants of the Portuguese "Vinho Verde" wine. 

These datasets can be viewed as classification or regression tasks. The classes are ordered and not balanced (e.g. there are many more normal wines than excellent or poor ones). Outlier detection algorithms could be used to detect the few excellent or poor wines. Also, we are not sure if all input variables are relevant. So it could be interesting to test feature selection methods.


**Attribute Information:**

Input variables (based on physicochemical tests):
1. fixed acidity
1. volatile acidity
1. citric acid
1. residual sugar
1. chlorides
1. free sulfur dioxide
1. total sulfur dioxide
1. density
1. pH
1. sulphates
1. alcohol

Output variable (based on sensory data):
12 - quality (score between 0 and 10)

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import scipy.stats as stats
import statsmodels.api as sm
```

```python
# url to csv file
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv'

# read csv file
data = pd.read_csv(url,sep=';')

# preview data set
data
```

```python
# preview charachteristics of dataset
data.describe()
```

```python
# print a list of unique values per feature
for i in data.keys():
  print(f"Column {i} has {len(np.unique(data[i]))} unique values.")
```

```python
# visulize dataset
fig,ax = plt.subplots(1,figsize=(18,5))
ax = sns.boxplot(data=data)
ax.set_xticklabels(ax.get_xticklabels(),rotation = 45)
plt.show()
```

```python
# as we see on plot some outliers above 250 in column 'total sulfur dixide'
# we will exclude values above 200 from dataset
data = data[ data['total sulfur dioxide'] < 200 ]

# let's build plot again and see changes
fig,ax = plt.subplots(1,figsize=(18,5))
ax = sns.boxplot(data=data)
ax.set_xticklabels(ax.get_xticklabels(),rotation = 45)
plt.show()
```

```python
# let's try to build some plots based on different columns combination
sns.scatterplot(x = data['residual sugar'],y = data['alcohol'])
plt.show()
```

```python
# pairwise plots
cols2plot = ['fixed acidity','citric acid','alcohol','pH','quality']
sns.pairplot(data[cols2plot],kind='reg',hue='quality')
plt.show()
```

```python
# t-test of volatile acidity on quality
# let's check what is the difference in values for
# volatile acidity for differnt type of wine quality

# extract data with quality 3
x = data['volatile acidity'][data['quality']==3]

# extract data with quality 8
y = data['volatile acidity'][data['quality']==8]

# run t-test
ttest = stats.ttest_ind(x,y)

plt.plot(np.random.randn(len(x))/30,x,'o',
         np.random.randn(len(y))/30+1,y,'o')
plt.xlim([-1,2])
plt.xticks([0,1],labels=['Qual 3','Qual 8'])
plt.ylabel('volatile acidity')
plt.title(f't={ttest[0]:.2f},p={ttest[1]:.5f}')
plt.show()
```

```python
# histogram plot shows number of times each quality metrics
# appears in data set
qualcounts = np.zeros(6)
n = 0

for i in range(3,9):
  qualcounts[n] = len(data[data['quality']==i])
  n += 1

plt.bar(range(3,9),qualcounts)
plt.xlabel('Quality rating')
plt.ylabel('Counts')
plt.show()
```

```python
# same logic as above but with pandas
counts = data['quality'].value_counts()

plt.bar(counts.keys(),counts)
plt.xlabel('Quality rating')
plt.ylabel('Counts')
plt.show()
```

```python
# check t test on quality=3/4 vs. 7/8
x = data['volatile acidity'][(data['quality']==3) | (data['quality']==4)]

# extract data with quality 8
y = data['volatile acidity'][(data['quality']==7) | (data['quality']==8)]

# run t-test
ttest = stats.ttest_ind(x,y)

plt.plot(np.random.randn(len(x))/30,x,'o',
         np.random.randn(len(y))/30+1,y,'o',
         markeredgecolor='k')
plt.xlim([-1,2])
plt.xticks([0,1],labels=['Qual 3+4','Qual 7+8'])
plt.ylabel('volatile acidity')
plt.title(f't={ttest[0]:.2f},p={ttest[1]:.5f}')
plt.show()
```

# Multiple regression

Prediction of the variability in 'Quality' measure using all features that are available in data set. So how well we can redict quality of wine based on all features. 

```python
# new data set for dependent variable 
dep_var = data['quality']
# new data set for independent variables
ind_var = data.drop(labels='quality',axis=1)

ind_var = sm.add_constant(ind_var)

# ordinary Least Squares is used to perform multiple regression 
# algorithm on data set
model = sm.OLS(dep_var,ind_var).fit()
print(model.summary())
```

```python
# get a list of sugnificant predictors of wine quality
sig_cols = list(model.pvalues[model.pvalues < .05].keys())
print("Significant predictors from the multiple regression:")
[ print(i) for i in sig_cols ];

# add column with quality to separate color according to quality 
sig_cols.append('quality')

# pairwise plots just for significant effects
sns.pairplot(data[sig_cols],kind='reg',hue='quality')
```

# Logistic regression

```python
# binarization of wine quality for usage of logistic regression
# let's find mean value of quality
binthresh = np.mean(data['quality'])

# build new dataset: old dataset plus new column of binary quality value  
data['binquality'] = data['quality']>binthresh
data
```

```python
# get 'charachteristical' columns that will be needed to build the model of logistical regression 
Xcols = []

for key in data.keys():
  if key not in ['quality','binquality']:
    Xcols.append(key)

Xcols
```

```python
model = sm.Logit(data['binquality'],data[Xcols])
results = model.fit(method = 'newton')
results.summary()
```

```python
# the same as in multiple regression logic, let's see the 
# list of characteristics with significant effect 
sig_colsL = list(results.pvalues[results.pvalues<.05].keys())

print("Significant predictors from the logistic regression:")
[ print(i) for i in sig_colsL ];

print('    ')

print("Significant predictors from the multiple regression:")
[ print(i) for i in sig_cols[:-1] ];
```
